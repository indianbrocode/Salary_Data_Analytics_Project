# Outlier Detection & Removal

# Visulaize Salary Outliers

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=Data['Salary'])
plt.title('Salary Outliers')
plt.show()
```

# IQR (Inter Quartile Range Method)

```python
Q1=Data['Salary'].quantile(0.25)
Q3=Data['Salary'].quantile(0.75)
IQR=Q3-Q1
```

```python
lower_limit=Q1-1.5*IQR
upper_limit=Q1+1.4*IQR
```

```python
Data = Data[(Data['Salary'] >= lower_limit) & (Data['Salary'] <= upper_limit)]
```

```python
Data
```

# Univariate Analysis

# Salary Distribition

```python
plt.figure(figsize=(8,5))
sns.histplot(Data['Salary'],bins=30,kde=True)
plt.ticklabel_format(style='plain',axis='x')
plt.title("Salary Distribtiton")
plt.xlabel("Salary")
plt.ylabel("count")
plt.show()
```

The above hist plot clearly shows the value of the graph in which the highest number of people is having lower to mid range Salary and very less people is having higest amount of Salary

## Mini Dashboard

```python
# Prepare Data
Rating_Salary=Data.groupby("Rating")["Salary"].mean().reset_index()
Data_Salary_by_City=Data.groupby('Location')['Salary'].mean().sort_values(ascending=False)

# Create Dashboard Layout

plt.figure(figsize=(14,10))

#1 Salary Distribution 
plt.subplot(2,2,1)
sns.histplot(Data["Salary"],bins=30,kde=True)
plt.title("Salary Distribution")

#2 Rating vs Salary
plt.subplot(2, 2, 2)
sns.scatterplot(x="Rating", y="Salary", data=Data, alpha=0.5)
sns.regplot(x="Rating", y="Salary", data=Data, scatter=False, color='red')
plt.title("Rating vs Salary")

#3 Average Salary vs Rating
plt.subplot(2, 2, 3)
sns.lineplot(x="Rating", y="Salary", data=Rating_Salary)
plt.title("Rating vs Salary")

# Top Locations
plt.subplot(2,2,4)
Data_Salary_by_City.sort_values().plot(kind='barh',color='green')
plt.title("Top 10 location by average Salary")

plt.suptitle("Salary Data Analysis Dashboard",fontsize=18, fontweight="bold")

plt.show()
```
